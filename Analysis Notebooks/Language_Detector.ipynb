```python
combined_test_data = test_data.copy()
combined_test_data['level_predictions_dan'] = test_data_baseline['dan_level_predictions']
combined_test_data['salary_predictions_dan'] = test_data_baseline['dan_salary_predictions']
```

```python
combined_test_data.head()
```

```python
def detect_language(text):
    try:
        return detect(text)
    except:
        return None


combined_test_data['language'] = combined_test_data['text'].apply(detect_language)
```

```python
language_names = {
    'en': 'English',
    'fr': 'French',
    'de': 'German',
    'nl': 'Dutch',
    'es': 'Spanish',
    'it': 'Italian',
    'pt': 'Portuguese',
    'pl': 'Polish',
    'da': 'Danish',
    'ro': 'Romanian'
}

combined_test_data['language_full_name'] = combined_test_data['language'].map(language_names)
```

```python
combined_test_data['language_full_name'].value_counts()
```

```python
combined_test_data.to_csv(r'test_data_lang_predictions.csv', index = False)
```
