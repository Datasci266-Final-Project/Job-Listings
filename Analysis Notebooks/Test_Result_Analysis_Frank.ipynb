```python
df_BERT_salary_match = np.where(test_data_3['salary_labels'] == test_data_3['salary_predictions_bert_base'])
df_Jobbert_salary_match = np.where(test_data_3['salary_labels'] == test_data_3['salary_predictions_jobbert'])
df_Roberta_salary_match = np.where(test_data_3['salary_labels'] == test_data_3['salary_predictions_xlm_roberta'])
df_DAN_salary_match = np.where(test_data_3['salary_labels'] == test_data_3['salary_predictions_dan'])

df_BERT_level_match = np.where(test_data_3['level_labels'] == test_data_3['level_predictions_bert_base'])
df_Jobbert_level_match = np.where(test_data_3['level_labels'] == test_data_3['level_predictions_jobbert'])
df_Roberta_level_match = np.where(test_data_3['level_labels'] == test_data_3['level_predictions_xlm_roberta'])
df_DAN_level_match = np.where(test_data_3['level_labels'] == test_data_3['level_predictions_dan'])
```

```python
print("salary test BERT - base: ", len(df_BERT_salary_match[0]))
print("salary test Jobbert: ", len(df_Jobbert_salary_match[0]))
print("salary test Roberta: ", len(df_Roberta_salary_match[0]))
print("salary test DAN: ", len(df_DAN_salary_match[0]))
print()
print("level test BERT - base: ", len(df_BERT_level_match[0]))
print("level test Jobbert: ", len(df_Jobbert_level_match[0]))
print("level test Roberta: ", len(df_Roberta_level_match[0]))
print("level test DAN: ", len(df_DAN_level_match[0]))
print()

print("salary test BERT - base: ", len(df_BERT_salary_match[0])/len_test_data_3)
print("salary test Jobbert: ", len(df_Jobbert_salary_match[0])/len_test_data_3)
print("salary test Roberta: ", len(df_Roberta_salary_match[0])/len_test_data_3)
print("salary test DAN: ", len(df_DAN_salary_match[0])/len_test_data_3)
print()
print("level test BERT - base: ", len(df_BERT_level_match[0])/len_test_data_3)
print("level test Jobbert: ", len(df_Jobbert_level_match[0])/len_test_data_3)
print("level test Roberta: ", len(df_Roberta_level_match[0])/len_test_data_3)
print("level test DAN: ", len(df_DAN_level_match[0])/len_test_data_3)
```
