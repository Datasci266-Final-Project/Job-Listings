```python
# Reverse Dictionaries
level_key = {
    0:"student_intern",
    1:"entry",
    2:"junior",
    3:"mid",
    4:"senior",
    5:"executive",
    -1:np.nan
}

#baseline salary
salary_key_baseline = {
    0:"<45k",
    1:"45-65k",
    2:"65-85k",
    3:"85-110k",
    4:"110-150k",
    5:"150-200k",
    6:">200k",
    -1:np.nan
}

#model 2+ salary
salary_key = {
    0:"<50k",
    1:"50-100k",
    2:"100-150k",
    3:"150-200k",
    4:">200k",
    -1:np.nan
}

```

```python
test_data['level_predictions_xlm_roberta'] = test_data['level_predictions_xlm_roberta'].map(level_key)
test_data['level_predictions_jobbert'] = test_data['level_predictions_jobbert'].map(level_key)
test_data['level_predictions_bert_base'] = test_data['level_predictions_bert_base'].map(level_key)
test_data['level_predictions_dan'] = test_data['level_predictions_dan'].map(level_key)
test_data['level_labels'] = test_data['level_labels'].map(level_key)


test_data['salary_predictions_xlm_roberta'] = test_data['salary_predictions_xlm_roberta'].map(salary_key)
test_data['salary_predictions_jobbert'] = test_data['salary_predictions_jobbert'].map(salary_key)
test_data['salary_predictions_bert_base'] = test_data['salary_predictions_bert_base'].map(salary_key)
test_data['salary_predictions_dan'] = test_data['salary_predictions_dan'].map(salary_key_baseline)
test_data['salary_labels'] = test_data['salary_labels'].map(salary_key)
test_data['salary_labels_baseline'] = test_data['salary_labels_baseline'].map(salary_key_baseline)
```
